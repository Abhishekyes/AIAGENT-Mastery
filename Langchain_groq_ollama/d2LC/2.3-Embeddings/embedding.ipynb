```python
import os
import subprocess
from langchain_community.embeddings import OllamaEmbeddings

# Disable tokenizers parallelism warning
os.environ["TOKENIZERS_PARALLELISM"] = "false"

# Step 1: Verify Ollama is installed and list available models
try:
    result = subprocess.run(["ollama", "list"], capture_output=True, text=True, check=True)
    models_output = result.stdout.strip()
    print("Ollama is installed. Available models:\n", models_output)
except Exception as e:
    print("Error running 'ollama list'. Ensure Ollama is installed and in your PATH.")
    print("Exception:", e)
    exit(1)

# Step 2: Parse model names from the output.
# Expected output format (tabular):
# NAME                ID              SIZE      MODIFIED
# gemma3:1b           2d27a774bc62    815 MB    14 minutes ago
# deepseek-r1:1.5b    a42b25d8c10a    1.1 GB    4 weeks ago
# llama3.2:1b         baf6a787fdff    1.3 GB    4 weeks ago

lines = models_output.splitlines()
models = []
header_found = False
for line in lines:
    # Look for header line containing "NAME" and "ID"
    if "NAME" in line and "ID" in line:
        header_found = True
        continue
    if header_found:
        parts = line.split()
        if parts:
            models.append(parts[0])
            
if not models:
    print("No models parsed from 'ollama list'. Please check the output format.")
    exit(1)

print("Parsed model names:", models)

# Step 3: Define a sample text for embedding generation.
sample_text = "This is a test document."

# Step 4: Iterate over each model, initialize embeddings, and generate the embedding.
for model in models:
    print("\n----------------------------------------")
    print("Initializing embeddings with model:", model)
    try:
        embeddings = OllamaEmbeddings(model=model)
        print("Initialized embeddings:", embeddings)
        print("Generating embedding for sample text:", sample_text)
        embedding_vector = embeddings.embed_query(sample_text)
        if embedding_vector is None:
            print(f"Model {model} returned None. It may not support embeddings or is misconfigured.")
        else:
            print(f"Embedding Vector for {model}:", embedding_vector)
    except Exception as e:
        print(f"Error processing model {model}:", e)
```
